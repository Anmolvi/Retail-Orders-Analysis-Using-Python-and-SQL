### Drive new columns discount, sale price and profit

```python
df1.head(2)

# new column selling price added
```

```python
df1["Profit"] = df1["Selling Price"] - df1["cost price"]
df1["Profit"]

# added new column as Profit by subtracting the selling price with cost price
```

```python
df1.head(2)
```

### Find month over month growth comparison for 2022 and 2023 sales eg: jan 2022 vs jan 2023

```python
df1.head(2)
```

```python
df1.groupby(["Month","Year"])
```

```python
a = df1.groupby(["Month","Year"]).sum("Total Sale")["Total Sale"].reset_index()
```

```python
a
```

```python
# Pivot table - inbuilt function in pandas
# index = rows
# aggfunc="sum" tells pandas: “If there are multiple sales in the same Month-Year, add them up.”

pivot_table = a.pivot_table(index = "Month", columns = "Year", values = "Total Sale", aggfunc="sum")
```

```python
pivot_table
```

```python
pivot_table.reset_index()
```

```python
# YOY_growth

pivot_table["YOY_growth"] = pivot_table[2022] - pivot_table[2023]
```

```python
pivot_table.reset_index()
```

### Which sub category had highest growth by profit in 2023 compare to 2022

```python
df1.head(2)
```

```python
c = df1.groupby(["Sub Category","Year"]).sum("Total Profit")["Total Profit"].reset_index()
```

```python
d = c.pivot_table(index = "Sub Category", columns = "Year", values = "Total Profit").reset_index()
d
```

```python
d["diff"] = d[2023] - d[2022]

d.sort_values("diff",ascending = False).head(1)
```
